```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
x=np.random.binomial(10,0.4,size=1000)#二项分布
xgrid = np.arange(0.5,15,1)
xcenter = (xgrid[1:]+xgrid[:-1])/2
fig = plt.figure()#图片大小
ax = fig.add_subplot(111)
ax.hist(x,xgrid)
fig.show()
print(np.mean(x),np.var(x))#平均值 10*0.4 方差为10*0.4*（1-0.4）
```

```python
x=np.random.poisson(10,size=1000)#泊松分布
xgrid = np.arange(0.5,30,1)
xcenter = (xgrid[1:]+xgrid[:-1])/2
fig=plt.figure()
ax=fig.add_subplot(111)
ax.hist(x,xgrid)
fig.show()
print(np.mean(x),np.var(x))#平均值与方差的期望均为n
```

```python
x=np.random.normal(1,4,size=1000)#正态分布
xgrid=np.arange(0.5,15,1)
xcenter=(xgrid[1:]+xgrid[:-1])/2
fig=plt.figure()
ax=fig.add_subplot(111)
ax.hist(x,xgrid)
print(np.mean(x),np.var(x))#函数中输入的是标准差
```

```python
x=np.random.chisquare(2,size=1000)
xgrid=np.arange(0.5,15,1)
xcenter=(xgrid[1:]+xgrid[:-1])/2
fig=plt.figure()
ax=fig.add_subplot(111)
ax.hist(x,xgrid)
print(np.mean(x),np.var(x))#平均值v方差为2v
```

```python
x=np.random.pareto(3,size=1000)
xgrid=np.arange(1,15,1)
xcenter=(xgrid[1:]+xgrid[:-1])/2
fig=plt.figure()
ax=fig.add_subplot(111)
ax.hist(x,xgrid)
print(np.mean(x),np.var(x))#方差与平均值有待商议
#help(np.random.pareto)
```
