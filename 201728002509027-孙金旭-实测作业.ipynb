```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fun(D,delta,msky):
    
    SNR=10
    mobj=np.arange(16,28,0.1)
    msun=5.48
    fsun=3.208*10**(-10)
    delam=95
    QE=0.5
    #msky=[19,21,23]
    #delta=[3,1,0.5,0.1]
    #D=[1,4,10]
    c=3*10**8
    h=6.626*10**-34
    lamda=4.4*10**-7
    plc=h*c/lamda
    fobj=fsun*10**((msun-mobj)/2.5)
    fsky=fsun*10**((msun-msky)/2.5)
    t=100*(fobj/plc+2*fsky/plc*3.14/4*delta**2)/((fobj/plc)**2*QE*3.14/4*D)**2
    return np.log10(t)
t1=fun(4,3,19)
t2=fun(4,3,21)
t3=fun(4,3,23)
t4=fun(4,1,19)
t5=fun(4,1,21)
t6=fun(4,1,23)
t7=fun(4,0.5,19)
t8=fun(4,0.5,21)
t9=fun(4,0.5,23)
t10=fun(4,0.1,19)
t11=fun(4,0.1,21)
t12=fun(4,0.1,23)
x=np.arange(16,28,0.1)
fig = plt.figure(figsize=[10,10])
ax = fig.add_subplot(111)

ax.plot(x,t1,label = 'see_disk =3,background = 19')
ax.plot(x,t2,label = 'see_disk =3,background = 21')
ax.plot(x,t3,label = 'see_disk =3,background = 23')
ax.plot(x,t4,label = 'see_disk =1,background = 19')
ax.plot(x,t5,label = 'see_disk =1,background = 21')
ax.plot(x,t6,label = 'see_disk =1,background = 23')
ax.plot(x,t7,label = 'see_disk =0.5,background = 19')
ax.plot(x,t8,label = 'see_disk =0.5,background = 21')
ax.plot(x,t9,label = 'see_disk =0.5,background = 23')
ax.plot(x,t10,label = 'see_disk =0.1,background = 19')
ax.plot(x,t11,label = 'see_disk =0.1,background = 21')  
ax.plot(x,t12,label = 'see_disk =0.1,background = 23')
plt.title("mag-time for D=4m")
plt.xlabel("mag_B")
plt.ylabel("log10(time/s)")
plt.legend(fontsize = 10)
plt.savefig("MT2.png")
fig.show()
            
```
